````python
L0 = 10.0   # Lunghezza iniziale
lam = 0.01  # Coefficiente lambda

def calcola_dilatazione():
    # Questa funzione è "cieca": cerca L0 e lam all'esterno.
    # Se cambiassimo nome alle variabili fuori non funzionerebbe
    L_finale = L0 * (1 + lam)

    # RETURN è fondamentale: restituisce il risultato al programma
    return L_finale

# La chiamo con le parentesi vuote () perché non accetta dati in ingresso
print(f"Lunghezza finale (fissa): {calcola_dilatazione()}")






# Ora definisco che la funzione DEVE ricevere due dati: L e coeff
def calcola_dilatazione(L, coeff):
    # Qui dentro uso i nomi generici L e coeff.
    # Non mi importa come si chiamano le variabili fuori.
    L_finale = L * (1 + coeff)
    return L_finale

# Posso passare le variabili esterne (L0, lam) dentro la funzione appena fatta
print(calcola_dilatazione(L0, lam))

# Oppure passo direttamente i numeri
print(f"Dilatazione Ferro: {calcola_dilatazione(4.0, 0.1)}")






# Posso anche definirli nella funzione
def calcola_dilatazione(L=5.0, coeff=0.1):
    L_finale = L * (1 + coeff)
    return L_finale

# CASO 1: Usa i default (5.0 e 0.1)
print(f"Standard: {calcola_dilatazione()}")

# CASO 2: Passo solo la lunghezza, Usa L=100 e coeff=0.1 (default)
print(f"Solo lunghezza: {calcola_dilatazione(L=100)}")

# CASO 3: Ignora i default e usa i miei dati
print(f"Tutto custom: {calcola_dilatazione(L=2.0, coeff=0.5)}")
````

## Loops and List

### Loop

Quelli più semplici sono il `while` e `for` loop.\
Il `while` loop ha la seguente struttura:
```python:
while condition: 
    <statement 1> 
    <statement 2> 
    ... 
<first statement after loop>
```

Il `for` è meno generale, e si applica a soli elementi di una **lista**:
```python:
for element in list:
    <statement 1>
    <statement 2>
    ...
<first statement after loop>
```

Posso usare la funzione `for` in combinazione con `range` per creare una lista di elementi partendo da una determinata funzione iterattiva

````python
P=100
r=5
N=10
amount=[]
for i in range(N+1):
    A= P* (1 + r/100)**i
    amount.append(A)
numeri= np.array(amount)
print(numeri.astype(int)) # converte tutto in numeri interi
````

### List
Le liste sono comode per creare delle concatenazioni di elementi; funzioni utili sono:
* `mylist[n]` per estrarre un elemento
* `mylist.append()` per aggiungere un elemento alla fine
* `del mylist[n]` per eliminare quell'elemento
* `a=b` crea la nuova variabile a che si riferisce a b, è un puntatore. 
* `a=b.copy()` crea una copia effettiva 

Se devo creare una lista di elementi posso usare la *list comprehension* ( `newlist = [expression for elemetns in the list]`) per scrivere in maniera più compatta.

````python
P=100
r=5
r_1= 3
N=10
amounts = [P* (1 + r/100)**i for i in range(N+1)]
amounts_1 = [P* (1 + r_1/100)**i for i in range(N+1)]
numeri= np.array(amount)
print(numeri.astype(int))
````

La funzione `zip` mi aiuta ad organizzare meglio quando devo presentare elementi di più liste, inoltre mi estra automaticamente i valori.

````python
for low, high in zip(amounts, amounts_1):
    print(int(low), int(high))
````

### Some type

````python
# list
# Sequence Type
# a collection of objects - mutable & ordered
['Ali','Xinyi','Miriam']

# tuple
# Sequence Type
# a collection of objects - immutable & ordered
('Thursday', 6, 9, 2018)

# set 
# a collection of unique objects - mutable & unordered
{'jack','sjoerd'}

# dict 
# Mapping Type
# mapping of key-value pairs
{'name':'DSAI', 'code':123,'credits':6}

# Use:
student = {"name": "Alice", "age": 20, "grade": "A"}
student_name = student["name"] # Access value by key.
print(student_name) # Alice
student["city"] = "New York" # Add new key-value pair.
print(student) 
student["age"] = 21 # Modify value by key.
print(student)
del student["grade"] # Remove key.
print(student)
````

### Pandas

````python
#Importiamo la libreria assegnandole il soprannome 'pd'
# Pandas per esempio serve a manipolare tabelle di dati in modo veloce.
import pandas as pd

# Creiamo un file (possiamo pure caricarlo esternamente da excel)
dati_fisica = {
    'punto': ['A', 'B', 'C'],
    'raggio_r': [0.02, 0.05, 0.08],
    'carica_Q': [1e-6, 1e-6, 1e-6]
}

df_creato = pd.DataFrame(dati_fisica)
df_creato.to_csv('dati_fisica.csv', index=False) # Salviamo il file CSV


# Carichiamo un file esterno.
# read_csv() legge il file e lo trasforma in una tabella Python chiamata 'contenuti'.
# il file deve trovarsi nella stessa cartella del Suo script.
contenuti = pd.read_csv('dati_fisica.csv')

# Calcoliamo la somma di tutti i valori nella colonna 'Carica_Q'
carica_totale = contenuti['carica_Q'].sum()

print("Ecco i dati che abbiamo letto:")
print(contenuti)
print(f"La somma vettoriale delle cariche è: {carica_totale} µC")

````

````python
import numpy as np                      # Importa la libreria NumPy e la rinomina 'np'

# --- Generazione numeri casuali ---
c = np.random.randint(5000, 10000)      # Genera UN solo numero intero casuale tra 5000 (incluso) e 10000 (escluso)
print(c)                                # Stampa il numero

c = np.random.randint(5000, 10000, 10)  # Il terzo numero (10) indica QUANTI numeri generare: crea un array di 10 interi
print(c)

c = np.mean(c)                          # Calcola la media aritmetica dell'array creato sopra usando la funzione np.mean()
print(c)

# A differenza delle liste, gli array NumPy devono contenere dati dello stesso tipo
dati = np.array([1, 2, 3, 4, 5, 6])     # Trasforma una lista Python [..] in un array NumPy

media = dati.mean()                     # Calcola la media chiamando il metodo direttamente sulla variabile 'dati'

print(media)

a = dati * 2 # posso compiere anche operazioni matematiche
print(a)
````
